```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math
import scipy.optimize as sci
import pandas as pd
```

```python
# CAPM line
def capm(rf, b, std):
    return rf + b* std
```

```python
# utility function
def utility(std, offset):
    return 0.05* (std+1)**4 + offset
```

```python
sns.set_style('dark')
steps = 100
std = np.linspace(0, 2, steps)
y_axis = 4.5
capm_line = []
utility_curve = []
utility_curve_second = []
utility_curve_third = []

for i in range(steps):
    capm_line.append(capm(rf=3, b=1.5, std=std[i]))
    utility_curve.append(utility(std=std[i], offset=3.72))
    #utility_curve_second.append(utility(std=std[i], offset=4.22))
    #utility_curve_third.append(utility(std=std[i], offset=3.22))


plt.figure(figsize=(10,8))
plt.plot(std, capm_line, color = 'black', label='CAPM line')
plt.plot(std, utility_curve, color = 'red', label='max. utility')
#plt.plot(std, utility_curve_second, color = 'red', alpha=0.2, linestyle='dashed')
#plt.plot(std, utility_curve_third, color = 'red', alpha=0.2, linestyle='dashed')
plt.hlines(y_axis, 0, 1, colors='red', linewidth = 0.8, linestyles='dashed', label="investor's optimal portfolio")
plt.vlines(1, 0, y_axis, colors='red', linewidth = 0.8, linestyles='dashed')
plt.xticks([])
plt.yticks([])
#plt.scatter(0, 3, color='red', zorder=5)
plt.text(-0.06, 2.98, r'$ \ r_{f}$', color='black', fontsize=14)
plt.text(-0.13, y_axis -0.02, r'$ \mu_{opt}$', color='black', fontsize=14)
plt.text(0.98, 1.85, r'$ \sigma_{opt}$', color='black', fontsize=14)
# market portfolio
plt.hlines(y_axis + 0.77, 0, 1.5, colors='black', linewidth = 0.8, linestyles='dashed', label='market portfolio')
plt.vlines(1.5, 0, y_axis +0.77, colors='black', linewidth = 0.8, linestyles='dashed')
plt.text(-0.13, y_axis +0.76, r'$ \mu_{m}$', color='black', fontsize=14)
plt.text(1.48, 1.85, r'$ \sigma_{m}$', color='black', fontsize=14)
plt.xlim(0, 2)
plt.ylim(2,8)
plt.xlabel('Risk ('+r'$ \sigma_{p}$'+')', loc='right', fontsize=16)
plt.ylabel('Return', loc='top', fontsize=16)
plt.title('Capital Market Line and Optimal Portfolio Choice', fontsize=20)
plt.legend(fontsize=14)
plt.show()
```
